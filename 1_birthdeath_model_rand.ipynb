```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Define Parameters
birth_rate = random.uniform(0.1, 0.5)  # Birth rate between 0.1 and 0.5
death_rate = random.uniform(0.1, 0.5)  # Death rate between 0.1 and 0.5

# Initialize Population Size
initial_population = 50  # Starting population size
max_time = 100  # Maximum time for simulation
```

```python
# Initialize variables
time = 0
population = initial_population
time_points = [time]
population_sizes = [population]

while time < max_time and population > 0:
    # Calculate rates of events
    birth_rate_total = birth_rate * population
    death_rate_total = death_rate * population
    total_rate = birth_rate_total + death_rate_total
    
    # Time to next event
    if total_rate == 0:
        break
    time_to_next_event = np.random.exponential(1 / total_rate)
    
    # Determine which event occurs
    if random.uniform(0, 1) < birth_rate_total / total_rate:
        population += 1  # Birth event
    else:
        population -= 1  # Death event
    
    # Update time
    time += time_to_next_event
    
    # Store results
    time_points.append(time)
    population_sizes.append(population)
```

```python
# Plot Results
plt.figure(figsize=(10, 6))
plt.step(time_points, population_sizes, where='post')
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.title('Birth-Death Process Simulation (Random)')
plt.grid(True)
plt.show()
```
